```python
da=d.drop('Restaurant', axis=1)
da
```

```python
da.corr()
```

```python
sns.heatmap(da.corr(), annot=True)
```

```python
x=da[['Extra Cheeze', 'Extra Mushroom', 'Extra Spicy', 'Price']]
x
```

```python
 y=da[' Size by Inch']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score,classification_report
logisreg = LogisticRegression()

#Train the model using Training Dataset
logisreg.fit(x_train, y_train)

# Prediction using test data
y_pred = logisreg.predict(x_test)
acc_logisreg = accuracy_score(y_test, y_pred)
print( 'Accuracy of Logistic Regression model : ', acc_logisreg )
```
