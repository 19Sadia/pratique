```python
plt.figure(figsize=(20,20), facecolor='skyblue')
plot=1

for columns in d:
    if plot<=9:
        ax=plt.subplot(3,3,plot)
        sns.histplot(d[columns], kde=True, color='r')
        plt.xlabel(columns,fontsize=20)
        plt.ylabel('count',fontsize=20)
    plot+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(20,20), facecolor='skyblue')
plot=1

for columns in d:
    if plot<=9:
        ax=plt.subplot(3,3,plot)
        sns.distplot(d[columns], hist=False)
        plt.xlabel(columns,fontsize=20)
        plt.ylabel('count',fontsize=20)
    plot+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(20,20), facecolor='pink')
plot=1

for column in d:
    if plot <=9:
        ax=plt.subplot(3,3,plot)
        sns.scatterplot(d[column])
        plt.xlabel(column, fontsize=20)
        plt.ylabel('count', fontsize=20)
    plot+=1
plt.tight_layout
```
