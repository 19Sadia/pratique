```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
```

### EDA

```python
sns.histplot(x='INDUS', kde=True, data=d)
```

```python
fig,ax=plt.subplots(12,2,figsize=(20,60))
m,n=0,0
for col in d:
    sns.histplot(d[col], kde=True, ax=ax[m][n])
    sns.distplot(d[col], hist=False, ax=ax[m][n+1])
    m+=1
    
```

```python
#oulier=crim,zn,b,dis,ptratio,age
```

```python
plt.figure(figsize=(20,60))
plot=1
for col in d:
    if plot<=15:
        ax=plt.subplot(12,2,plot)
        sns.histplot(d[col],kde=True)
        plt.xlabel(col,fontsize=20)
        plt.ylabel('count',fontsize=20)
    plot+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(20,60))
plot=1
for col in d:
    if plot<=15:
        ax=plt.subplot(12,2,plot)
        sns.scatterplot(d[col])
        plt.xlabel(col,fontsize=20)
        plt.ylabel('count',fontsize=20)
    plot+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(20,60))
plot=1
for col in d:
    if plot<=15:
        ax=plt.subplot(12,2,plot)
        sns.boxplot(d[col])
        plt.xlabel(col,fontsize=20)
        plt.ylabel('count',fontsize=20)
    plot+=1
plt.tight_layout()
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
dat[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
         'PTRATIO', 'B', 'LSTAT', 'MEDV']]=mm.fit_transform(dat[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT', 'MEDV']])
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(x.corr(),annot=True)
```

```python
y=dat['MEDV']
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2=r2_score(y_test,y_predict)
r2
```
